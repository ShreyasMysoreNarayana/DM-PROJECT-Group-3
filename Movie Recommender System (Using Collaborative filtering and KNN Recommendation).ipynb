# KNN Recommendation

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
# This takes any choice, or else we can give any userId or the movie name itself.

query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
movie_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```

# Prediction

```python
# List of movies provided as input
input_movies = ['Elizabeth (1998)', 'Exotica (1994)', 'Sense and Sensibility (1995)', 'American Beauty (1999)', 'Bringing Out the Dead (1999)']
# Assuming `movie_features_df` has movie titles as its index
selected_movies = movie_features_df.index.intersection(input_movies)

# Aggregate the features for the input movies
aggregated_features = movie_features_df.loc[selected_movies].mean(axis=0)

# Find recommendations using the aggregated features
distances, indices = model_knn.kneighbors(aggregated_features.values.reshape(1, -1), n_neighbors=200)

# Filter out the input movies from the recommendations
recommendations = [
    (movie_features_df.index[indices.flatten()[i]], distances.flatten()[i])
    for i in range(len(indices.flatten()))
    if movie_features_df.index[indices.flatten()[i]] not in selected_movies
]

# Limit to the top 100 unique recommendations
recommendations = recommendations[:100]

# Display the recommendations
print(f"Recommendations based on the provided movies:\n")
for i, (movie_id, distance) in enumerate(recommendations, start=1):
    print(f"{i}: {movie_id}, with distance of {distance}")
```

```python
recommended_movies_list = [movie_id for movie_id, distance in recommendations]
```

## Collaborative Filtering and Association rules 

### 1. Prepare Data for Association Rule Mining

```python
# Group movies by user
transactions = popular_movies_df.groupby('userId')['title'].apply(list).reset_index(drop=True)

print(transactions.head())  # Displays the list of movies per user
```

```python
transactions
```

### 2. Transform Data Using One-Hot Encoding

```python
from mlxtend.preprocessing import TransactionEncoder

# Convert transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

# Create a DataFrame from the one-hot encoded array
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df_encoded
```

### 3. Apply Apriori Algorithm

```python
from mlxtend.frequent_patterns import fpgrowth

# Use FP-Growth to find frequent itemsets
frequent_itemsets = fpgrowth(df_encoded, min_support=0.02, use_colnames=True)

print(frequent_itemsets.head())  # Displays frequent itemsets
```

### 4. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```
